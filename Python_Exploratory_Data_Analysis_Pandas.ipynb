```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib


#look at the data patterns, relationships, any outliers/areas to CLEAN
df = pd.read_csv(r"C:\Users\Ezhan Khan\Documents\PYTHON PRACTICE\PYTHON for DATA ANALYTICS\Pandas_Data_Cleaning_and_Analysis\world_population.csv")
df
#dataset contains information on Countries Populations over the years
#Can change the Number of 'Decimal Places' for FLOAT Columns in the DataFrame to '2dp':
pd.set_option('display.float_format', lambda x: '%.2f'% x)
df
```

```python
#Inspect the Columns, see if there are any NULL/Missing values:
df.info()
#have a few missing values for some of the Population Columns. 
```

```python
#get a quick view of the Summary Statistics for the Columns
df.describe(include="all")
```

```python
#View the Null Values in the DataFrame:
df.isnull().sum()
#IMPORTANT to know how much missing data EACH Column CONTAINS!
```

```python
#Another way to view Nulls as PERCENTAGE of TOTAL:
for col in df.columns:
    pct_missing = np.round(np.mean(df[col].isnull()),2)  #find missing data
    print(f'{col} - {pct_missing}%')  #for each column, as f-string
```

```python
#Viewing Number of UNIQUE Values within EACH Column
df.nunique()
```

```python
#Viewing Top 10 Countries with 'Largest' Populations in 2022 (by SORTING):
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
#Bottom 10 'Smallest' Populations in 2022:
df.sort_values(by="2022 Population", ascending=True).head(10)
```

```python
#Sorting by 'World Population Percentage':
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
#Look at CORRELATIONS between NUMERIC COLUMNS:
df.corr()  
#this is much clearer if VISUALIZED as a HEAT MAP:
sns.heatmap(df.corr(), annot=True)
matplotlib.rcParams['figure.figsize'] = (20,8) 
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")  
plt.show()
#now, is clearly colour-coded, where lighter colour = HIGH Correlation
#see how 'Populations' in Different Years are HIGHLY Correlated with each other!
#'World Population Percentage' is also HIGHLY Correlated with Country Populations
#SOME correlation with 'Area' (Larger Area = Larger Population)

#Viewing Correlation can be HELPFUL when deciding what Variables to VISUALIZE and Analyze LATER
```

```python
#Have Certain 'Continents' grown faster than others and in what way?

# GROUP by 'Continent' Column:
df.groupby('Continent').mean()
#What Countries which are in 'Oceania' Continent?
df[df['Continent'].str.contains('Oceania')]      #or, df['Continent']=='Oceania' 
```

```python
#Sort GROUPED Continents BY '2022 Population' (most recent):
df2 = df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
df2
#as expected, 'Asia' has highest average population, each country makes up roughly 1.1% of world's population
```

```python
#Now, can VISUALIZE this Data. BUT, in Current WIDE DATA FORMAT, is NOT what we want!

#Before visualizing, best to 'TRANSPOSE' DATA (FLIP around)
#First, FILTER to JUST INCLUDE 'Population' COLUMNS (only need these!)
#(may need to REORDER these, so goes from '1970 - 2022')
df2 = df2[::-1]          #Used '[::-1]' to REVERSE ORDER of 'Population' Columns!
df3 = df2.transpose()
df3

df3.plot(kind='line', figsize = (12,10))
#Really interesting trends shown! Asia consistently has SIGNIFICANTLY Higher Population over this time-span!

```

```python
#Viewing 'OUTLIERS' in the Columns - easily done with BOXPLOTS: 
df.boxplot(figsize=(20,10))
#Outliers are dots OUTSIDE of 'Box and Whiskers'.
#since we are dealing with 'Population' data here, this is acceptable
#(but IF dealing with revenue/sales data, can be MORE IMPORTANT to CONSIDER WHY these Outliers may exist!)
```
